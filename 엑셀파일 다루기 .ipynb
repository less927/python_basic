```python
import pandas as pd

excel_exam_data1 = {'학생' : ['A', 'B', 'C', 'D', 'E', 'F'],
                    '국어' : [80, 90, 95, 70, 75, 85],
                    '영어' : [90, 95, 70, 85, 90, 95,],
                    '수학' : [85, 95, 75, 80, 85, 100]}

df1 = pd.DataFrame(excel_exam_data1,columns=['학생', '국어', '영어','수학'])
df1
```

```python
excel_writer = pd.ExcelWriter("H:/code/python basic/python_basic/data/학생시험성적2.xlsx", engine='xlsxwriter')
df1.to_excel(excel_writer, index = False)
excel_writer.close()
```

```python
excel_writer2 = pd.ExcelWriter("H:/code/python basic/python_basic/data/학생시험성적3.xlsx", engine='xlsxwriter')
df1.to_excel(excel_writer2, index = False, sheet_name='중간고사')
excel_writer2.close()
```
