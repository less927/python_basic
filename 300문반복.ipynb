```python
#for문의 실행결과를 예측하라.
과일 = ["사과", "귤", "수박"]
for 변수 in 과일:
    print(변수)
```

```python
과일 = ["사과", "귤", "수박"]
for 변수 in 과일:
  print("#####")
  #기초문법 배우기-1에서 for문의 핵심은 "들여쓰기된 코드가 자료구조에 저장된 데이터 개수만큼 반복된다"라고 설명했습니다.
  #`과일 = ["사과", "귤", "수박"]` 에는 세 개의 데이터가 저장돼 있으므로 들여쓰기된 `print("####")`코드가 세 번 실행됩니다.
```

```python
리스트 = [13, 21, 12, 14, 30, 18]
#리스트에서 20 보다 작은 3의 배수를 출력하라

for nums in 리스트:
  if (nums % 3==0) and (nums < 20):
    print (nums)
```
