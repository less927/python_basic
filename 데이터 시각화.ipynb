```python
#그래프 예시
import matplotlib.pyplot as plt
import numpy as np

# 데이터 생성
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x)

# 그림 생성과 서브 플롯 할당
fig, axs = plt.subplots(2, 2)

# 각 서브 플롯에 데이터 그리기
axs[0, 0].plot(x, y1)
axs[0, 0].set_title('Sine Function')

axs[0, 1].plot(x, y2)
axs[0, 1].set_title('Cosine Function')

axs[1, 0].plot(x, y3)
axs[1, 0].set_title('Tangent Function')

axs[1, 1].plot(x, y4)
axs[1, 1].set_title('Exponential Function')

plt.tight_layout()  # 서브 플롯 간의 간격을 조정하여 겹치지 않도록 함
plt.show()
```

```python
members_IDs = ['m01', 'm02', 'm03', 'm04' ]
before_ex = [27, 35, 40, 33]
after_ex = [30, 38, 42, 37]
```

```python
import matplotlib.pyplot as plt
import numpy

n_data = len (members_IDs)
index = np.arange(n_data)
plt.bar(index, before_ex)
plt.show()
```

```python
plt.bar(index, before_ex, tick_label = members_IDs)
plt.show()
```

```python
colors=['r', 'g', 'b', 'm']
plt.bar(index, before_ex, color=colors, tick_label = members_IDs)
plt.show()
```

```python
plt.bar(index, before_ex, tick_label = members_IDs, width = 0.6)
plt.show()
```

```python
colors=['r', 'g', 'b', 'm']
plt.barh(index, before_ex, color=colors, tick_label = members_IDs)
plt.show()
```

```python
barWidth = 0.4
plt.bar(index, before_ex , color='c', align='edge', width = barWidth, label='before')
plt.bar(index + barWidth, after_ex, color='m', align='edge', width = barWidth, label='after')
        
plt.xticks(index + barWidth, members_IDs)
plt.legend()
plt.xlabel('회원id')
plt.ylabel('윗몸일으키기 횟수')
plt.title('운동 전후 복근비교')
plt.show()
```
