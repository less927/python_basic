```python

x = np.arange(-5, 5, 0.1)
y1 = x**2 - 2
y2 = 20 * np.cos(x)**2

# 첫 번째 그림과 서브플롯 생성
fig1, ax1 = plt.subplots()
ax1.plot(x, y1)

# 두 번째 그림과 서브플롯 생성
fig2, ax2 = plt.subplots()
ax2.plot(x, y2)

# 첫 번째 그림의 서브플롯에 y2 그래프 추가
ax1.plot(x, y2)

# 두 번째 그림의 서브플롯에 y1 그래프 추가
ax2.plot(x, y1)

plt.show()
```

```python
#그래프 예시
import matplotlib.pyplot as plt
import numpy as np

# 데이터 생성
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x)

# 그림 생성과 서브 플롯 할당
fig, axs = plt.subplots(2, 2)

# 각 서브 플롯에 데이터 그리기
axs[0, 0].plot(x, y1)
axs[0, 0].set_title('Sine Function')

axs[0, 1].plot(x, y2)
axs[0, 1].set_title('Cosine Function')

axs[1, 0].plot(x, y3)
axs[1, 0].set_title('Tangent Function')

axs[1, 1].plot(x, y4)
axs[1, 1].set_title('Exponential Function')

plt.tight_layout()  # 서브 플롯 간의 간격을 조정하여 겹치지 않도록 함
plt.show()
```

```python
city = ['서울', '인천', '대전', '대구', '울산', '부산', '광주',]
#위도와 경도
lat = [37.56, 37.451, 36.35, 35.87, 35.53, 35.18, 35.16]
lon = [126.97, 126.70, 127.38, 128.60, 129.31, 129.07, 126.85]

pop_den = [16154, 2751, 2839, 2790, 1099, 4454, 2995]
size = np.array(pop_den) * 0.2
colors = ['r', 'g', 'b', 'c', 'm', 'k', 'y']

plt.scatter(lon, lat, s=size, c=colors, alpha=0.5)
plt.xlabel('경도')
plt.ylabel('위도')
plt.title('지역별 인구밀도')

for x, y, name in zip(lon, lat, city):
    plt.text(x, y, name) #위도 경도에 맞게 도시이름 출력
    
plt.show()    
```

```python
members_IDs = ['m01', 'm02', 'm03', 'm04' ]
before_ex = [27, 35, 40, 33]
after_ex = [30, 38, 42, 37]
```

```python
import matplotlib.pyplot as plt
import numpy

n_data = len (members_IDs)
index = np.arange(n_data)
plt.bar(index, before_ex)
plt.show()
```

```python
plt.bar(index, before_ex, tick_label = members_IDs)
plt.show()
```

```python
colors=['r', 'g', 'b', 'm']
plt.bar(index, before_ex, color=colors, tick_label = members_IDs)
plt.show()
```

```python
plt.bar(index, before_ex, tick_label = members_IDs, width = 0.6)
plt.show()
```

```python
colors=['r', 'g', 'b', 'm']
plt.barh(index, before_ex, color=colors, tick_label = members_IDs)
plt.show()
```

```python
barWidth = 0.4
plt.bar(index, before_ex , color='c', align='edge', width = barWidth, label='before')
plt.bar(index + barWidth, after_ex, color='m', align='edge', width = barWidth, label='after')
        
plt.xticks(index + barWidth, members_IDs)
plt.legend()
plt.xlabel('회원id')
plt.ylabel('윗몸일으키기 횟수')
plt.title('운동 전후 복근비교')
plt.show()
```
